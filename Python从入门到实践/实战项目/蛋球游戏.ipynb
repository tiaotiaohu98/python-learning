```python
import time
from tkinter import *
import math
from random import *


# 事件触发函数
def move_boll():
    print('游戏开始了!')
    canvas.delete("all")
    tk.update()
    # canvas.delete(id1)
    # 创建小球并使小球移动
    text = "你真是个小可爱!!!"
    id1 = canvas.create_text(200, 100, text=text, font=('宋体', 20), fill='red')
    id2 = canvas.create_oval(10, 50, 60, 100, fill='yellow', outline='lightgray')
    for x in range(180):
        canvas.move(id1, 5, 0)
        canvas.move(id2, 0, 5)
        tk.update()  # 小球移动后,强制tkinter重新绘制
        time.sleep(0.05)  # 每0.05秒移动一次进行下一次循环


# 背景图滚动
# def bgUpdate(source):
#     # 更改画布颜色
#     red = rSlider.get()
#     green = gSlider.get()
#     blue = bSlider.get()
#     print("R=%d,G=%d,B=%d" % (red, green, blue))
#     myColor = "#%02x%02x%02x" % (red, green, blue)
#     canvas.config(bg=myColor)


# 创建tkinter对象
tk = Tk()

# 为TKINTER窗口添加标题
tk.title('蛋球游戏')

# 建立画布
canvas = Canvas(tk, width=900, height=580, bg='black')
canvas.pack()

# # 创建一个按钮，并绑定按钮事件函数
# 按钮使小球重新移动
startGame = '重新开始游戏'
start_btn = Button(tk, text=startGame, command=move_boll)
start_btn.pack()


# 设置背景色
# rSlider = Scale(tk, from_=0, to=255, command=bgUpdate)
# gSlider = Scale(tk, from_=0, to=255, command=bgUpdate)
# bSlider = Scale(tk, from_=0, to=255, command=bgUpdate)

# gSlider.set(10)
# rSlider.grid(row=1, column=1)
# gSlider.grid(row=1, column=2)
# bSlider.grid(row=1, column=3)
# canvas.grid(row=2, column=1, columnspan=3)

def ball_move_direction(event):
    if event.keysym == 'Left':
        canvas.move(1, -50, 0)
    if event.keysym == 'Right':
        canvas.move(1, 50, 0)
    if event.keysym == 'Up':
        canvas.move(1, 0, -50)
    if event.keysym == 'Down':
        canvas.move(1, 0, 50)


canvas.create_oval(225, 125, 275, 175, fill='red')
canvas.bind_all('<KeyPress-Left>', ball_move_direction)
canvas.bind_all('<KeyPress-Right>', ball_move_direction)
canvas.bind_all('<KeyPress-Up>', ball_move_direction)
canvas.bind_all('<KeyPress-Down>', ball_move_direction)

# 保证程序不停止，直到用户点击窗口关闭按钮
mainloop()
```
