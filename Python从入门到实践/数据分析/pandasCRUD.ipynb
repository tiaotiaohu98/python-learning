```python
import pandas as np
import pandas as pd

df = pd.read_csv('data/1960-2019全球GDP数据.csv',encoding='gbk')
df = df.head()
# df

df1 = df.copy()
# df1

# 新增列数据数量必须和行数相等
df1['new col 1'] = 33

df1['new col 2'] = [1,2,3,4,5]
df1['new col 3'] = df1.year * 2

print(df)
print(df1)
```

```python
# 增加数据
# 1、新增列，一组数据的数据个数必须和行数保持一致
# print(df)
df1 = df1.assign(new0=66)
# df.assign(new1=[1,2,3,4])
df1 = df1.assign(new1=[1,2,3,4,5])

# 2、新列名=Series对象 该对象的索引和df的索引要一致
s = pd.Series([1,2,3,4]) # 缺少的字段填充为NaN
df1 = df1.assign(new2=s)
df1 = df1.assign(new3=df1.year+df1.GDP)
# df1

# 3、新列名 = 自定义函数名
def foo(df):
    print('='*10)
    print(df)
    print('='*5 + '上面输出的是传入的df')
    ret = df.index.values
    
    return ret
# 新列的值按函数返回值填充
df1 = df1.assign(new4=foo)
print(df1)

# 同时添加多列
def foo1(df):
    return 22

def foo2(df):
    return df.year+1

df1 = df1.assign(new5=foo1,
           new6=foo2)

print(df1)
```

```python
# 删除
# 删除行数据
# print(df1)
# df1.drop([0])
# df1.drop([0,2,4])
# 对某一个Series对象按索引删除
print(df1.GDP.drop([0, 2]))

# 删除列数据 , 指定参数axis=1(0为纵轴，1为横轴),则按列删除 
df1.drop(['new col 2'], axis=1)
print(df1)

# del删除指定列
#     del是直接永久删除原df中的列
#     drop是返回删除后的df或Series, 原df或series没有被改变
del df1['new col 3']
df1
```

```python
# 去重
# 添加一部分重复数据
df2 = df1._append(df1).reset_index(drop=True) # type: ignore
print(df2)

# 去除重复数据
df2.drop_duplicates()

# Series去重
# df2.country.drop_duplicates()
df2.country.unique()
```

```python
df7 = pd.DataFrame({
    '姓名':['小明', '小美', '小强', '小兰'],
    '成绩':[100, 90, 90, 80]
})

print(df7.rank())

# 值相同，按最小值排序，排名不连续
df7['成绩排名']=df7['成绩'].rank(method='min',ascending=False)
print(df7)

# 值相同，按最大值排序，排名不连续
df7['成绩排名']=df7['成绩'].rank(method='max',ascending=False)
print(df7)

# 值相同，排名连续
df7['成绩排名']=df7['成绩'].rank(method='dense',ascending=False)
df7
```

```python
# 聚合函数 max()  min()   mean()   sum()   count()
print(df7.min())
print(df7.max())
print(df7['成绩'].mean())
print(df7.sum())
print(df7.count())
```

```python
# 分组
df = pd.read_csv('data/uniqlo.csv')

# 基于一列进行分组
gender = df.groupby(['gender_group'])
print(gender)
gender['city']

# 基于多列进行分组
gs = df.groupby(['gender_group','channel'])
# gs

# 获取分组后数据
# print(gs.first())  # 取出每组第一条数据
gs.last()  # 取出每组最后一条数据

# 取其中一组数据
gs.get_group(('Female','线上'))
```

```python
# 分组聚合
df.groupby(['city','channel']).agg({'revenue':'mean',
                                     'unit_cost':'sum'})

# 分组过滤
# df.groupby(['city']).filter(lambda x: x['revenue'].sum() > 1000)
df.groupby(['city'])['revenue'].filter(lambda x: x.mean() > 200)
```
