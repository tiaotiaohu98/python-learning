```python
import pandas as pd


df = pd.read_csv('data/movie_top_250.csv')
df.rename(columns={'标题':'title', '标题链接': 'url', 'bd':'director', '数量':'grade', '编号': 'evaluation','inq':'description'}, inplace=True)
df['director'] = df.director.apply(lambda x: x.split('导演:',1)[-1].strip())
df['director'] = df.director.apply(lambda x: x.split('主演:',1)[0].strip())
df
```

```python
# 词云图
from pyecharts import options as opts
from pyecharts.charts import WordCloud


titles = df.title.to_list()
evalutrions = df.evaluation.str.rstrip('人评价').astype('int').to_list()

data = list(zip(titles, evalutrions))

(
    WordCloud()
    .add(series_name="豆瓣电影Top250分析", data_pair=data,word_size_range=[6,66])
    .set_global_opts(title_opts=opts.TitleOpts(
        title="豆瓣电影Top250分析",
        title_textstyle_opts=opts.TextStyleOpts(font_size=23)),
        tooltip_opts=opts.TooltipOpts(is_show=True),
        )
).render_notebook()
```

```python
# 柱形图
# 分组统计
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker


grades = df.groupby(['grade'])['title'].count().index.to_list()
# grades
# 获取数据
nums = df.groupby(['grade'])['title'].count().values.tolist()
# nums

(
    Bar()
   .add_xaxis(grades)
   .add_yaxis('电影评分', nums)
   .set_global_opts(title_opts=opts.TitleOpts(title='豆瓣电影Top250评分', subtitle='不同评分对应的电影数量分析'))
).render_notebook()
```
