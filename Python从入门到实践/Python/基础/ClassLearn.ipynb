```python
#创建类
class Dog:
    '''模拟小狗实验'''

    def __init__(self,name,age):
        self.name= name
        self.age=age
        self.master="1"

    def sit(self):
        print(f"{self.name} is now sitting.")

    def roll_over(self):
        print(f"{self.name} rolled over!")

# 创建实例
mydog= Dog("whillie",6)

# 调用类属性
print(f'My dog name is {mydog.name}')
print(f'My dog is {mydog.age} years old')

# 调用类方法
mydog.sit()
mydog.roll_over()

your_dog= Dog("lucy", 3)

print(your_dog.name,your_dog.age)
your_dog.sit()
your_dog.roll_over()

print(mydog.master)
```
