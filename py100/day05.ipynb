```python
# 列表推导式
list1 = []
for i in range(10):
    list1.append(i)
print(list1)

list2 = [i for i in range(10)]
print(list2)


# 列表推导式生成0到9 之间的偶数
list1 = []
for i in range(10):
    if i%2==0:
        list1.append(i)
print(list1)

list2 = [i for i in range(10)]
print(list2)

list3 = []
for i in range(2):
    for j in range(3):
        list3.append((i,j))
print(list3)

list4 = [(i,j) for i in range(2) for j in range(3)]
print(list4)
        
```

```python
# 不定长参数
# 关键字*args 
# 数据类型为元组类型，传参时只能用位置传递
def func_1(*args):
    print(args)
    print(type(args))
    
func_1(1,2,3)

# 关键字 **kwargs
# 数据类型为字典，传参时只能用关键字传参
def func_2(**kwargs):
    for k,v in kwargs.items():
        print(k,v)
    print(type(kwargs))
    
func_2()
func_2(name='zy')
func_2(name='zy',age=25)
```

```python
# 不定长参数混合使用
# *args,**kwargs 混合使用时,*arg必须在前面,**kwargs必须在后面
def func3(*args,**kwargs):
    sum = 0
    for i in args:
        sum += i
    for j in kwargs.values():
        sum += j
    print(sum)
    
list1=[1,2,3]
dict1={'a':4,'b':5}

func3(*list1,**dict1)
```

```python
# lamda表达式
def func4():
    return 100

print(func4())

func5 = lambda :100
print(func5())

def func6(a,b,c=30):
    return a+b+c

func61 = lambda a,b,c=30:a+b+c
print(func61(10,20))

func7 = lambda *args:args
print(func7(1,2,3))

func8 = lambda **kwargs:kwargs
print(func8(a=10,b=20,c=30))
```
